```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("ML470_S1_HR_Data_Practice.xlsx")

attrition_counts = df.groupby(['salary', 'left']).size().unstack()

attrition_counts.plot(kind='bar')

plt.title("Employee Attrition by Salary Level")
plt.xlabel("Salary Level")
plt.ylabel("Number of Employees")
plt.legend(["Stayed", "Left"],loc='upper right')
plt.tight_layout()
plt.show()
```
