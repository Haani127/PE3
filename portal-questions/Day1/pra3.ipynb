```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("ML470_S1_HR_Data_Practice.xlsx")

dept_attrition = df.groupby(['Department', 'left']).size().unstack()

dept_attrition.plot(kind='bar')

plt.title("Department-wise Employee Retention")
plt.xlabel("Department")
plt.ylabel("Number of Employees")
plt.legend(["Stayed", "Left"], loc="upper right")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
