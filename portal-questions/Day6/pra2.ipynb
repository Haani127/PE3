```python
import pandas as pd
import warnings
import matplotlib.pyplot as plt
import seaborn as sns

def assess_outliers(data1):
    plt.figure(figsize=(14, 6))
    sns.boxplot(data=data1)
    plt.xticks(rotation=90)
    plt.title("Outlier Assessment Using Boxplots")
    plt.show()

def check_correlation_plot(input_df):
    plt.figure(figsize=(12, 8))
    corr = input_df.corr()
    sns.heatmap(corr, annot=True, cmap="YlGnBu", fmt=".2f")
    plt.title("Feature Correlation Analysis Using Heatmap")
    plt.show()

warnings.filterwarnings("ignore")

mobile_df = pd.read_excel("ML470_S6_Mobile_Data_Practice.xlsx")

X = mobile_df.loc[:, mobile_df.columns != 'price_range']

assess_outliers(X)
check_correlation_plot(X)
```
